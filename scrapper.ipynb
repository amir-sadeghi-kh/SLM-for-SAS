```python
import requests
from bs4 import BeautifulSoup
import json

# Step 1: Scrape Web Pages
def fetch_page(url):
    response = requests.get(url)
    response.raise_for_status()  # Check if the request was successful
    return response.text

def extract_sas_code(html_content):
    soup = BeautifulSoup(html_content, 'html.parser')
    # Example: Assuming SAS code is inside <pre> tags with class "xisDoc-code"
    code_blocks = soup.find_all('pre', class_='xisDoc-code')
    codes = [block.get_text() for block in code_blocks]
    print(codes)
    return codes

def scrape_urls(urls):
    all_codes = []
    for url in urls:
        print(f"Scraping {url}...")
        html_content = fetch_page(url)
        print(html_content)
        codes = extract_sas_code(html_content)
        all_codes.extend(codes)
    return all_codes

# Step 2: Assign Tags (Example: Tags based on keywords or patterns)
def assign_tags(codes):
    tagged_data = []
    for code in codes:
        if 'proc' in code:
            tags = ['proc']
        elif 'data' in code:
            tags = ['data']
        else:
            tags = ['general']

        tagged_data.append({'code': code, 'tags': tags})
    return tagged_data

# Step 3: Save Data for Fine-Tuning
def save_data(data, filename='sas_codes.json'):
    with open(filename, 'w') as file:
        json.dump(data, file, indent=2)

# Example URLs to scrape
urls = [
    'https://go.documentation.sas.com/doc/en/pgmsascdc/9.4_3.4/ormpug/ormpug_optmodel_examples01.htm'
]

# Run the scraping and processing
codes = scrape_urls(urls)
tagged_codes = assign_tags(codes)
save_data(tagged_codes)

print(f"Data has been saved to sas_codes.json")

# Note: Ensure to install required libraries before running the script
# pip install requests beautifulsoup4
```
